**Chest Pain Type = Tipo de dor no peito**

```python
data['ChestPainType'].value_counts()
```

ASY = Assintomático

NAP = Dor não Anginosa

ATA = Angina Atípica

TA = Angina Típica

Angina = Dor no peito causada pela diminuição do fluxo de sangue no coração

```python
sns.countplot(x='ChestPainType', data=data, palette='pastel', hue='ChestPainType', legend=False);
```

Resting BP (Blood Pressure) = pressão sanguínea em repouso em mmHg (sistólica)

```python
data.RestingBP.value_counts().sort_index()
```

```python
sns.histplot(data, x='RestingBP', bins=30, color='red', kde=True, stat='count');
```

**Cholesterol = colesterol sérico (mg/dl)**


```python
data.Cholesterol.value_counts().sort_index()
```

```python
sns.histplot(data, x='Cholesterol', bins=30, color='blue', kde=True, stat='count');
```

**Fasting BS (Blood sugar) = açúcar no sangue em jejum (mg/dl)**

```python
data.FastingBS.value_counts()
```

0: Fasting BS < 120 md/dl (não diabético)

1: Fasting BS >= 120 md/dl (diabético)

```python
sns.countplot (x='FastingBS', data=data, palette='pastel', hue='FastingBS', legend=False);
```

**Resting ECG = eletrocardiograma em repouso**

```python
data.RestingECG.value_counts()
```

Normal = normal

LVH = Hipertrofia ventricular esquerda

ST = Anormalidade da onda ST-T Segmento ST: Qualquer desnível do segmento ST, para cima (supradesnível) ou para baixo (infradesnível) deve ser avaliado pois pode corresponder a um infarto agudo no miocárdio

```python
plt.figure(figsize=(8,5))
sns.countplot(x='RestingECG', hue='RestingECG', data=data, palette='pastel', legend=False)
plt.show()
```
